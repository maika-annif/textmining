<h1>Python Basics 2</h1>
<strong>Themen: Kontrollstrukturen (Fallunterscheidungen, Schleifen und einfache Funktionen)</strong>
<p>
In einem Computerprogramm werden Befehle sequenziell, d.h. der Reihe nach, Befehl für 
Befehl, abgearbeitet. Jede Zeile entspricht einem Befehl. Im Unterschied zu vielen anderen Programmiersprachen wird ein Befehl nicht(!) mit einem Semikolon abgeschlossen.
</p>
<p>  
Kontrollstrukturen wie Fallunterscheidungen und Schleifen verändern den sequenziellen Programmlauf. 
Der Code wird dazu in Anweisungsblöcke gegliedert. In Python wird ein Anweisungsblock durch einen Doppelpunkt und Einrückungen gekennzeichnet. 
</p>
<p> 
Die Ausführung von Anweisungs-Blöcken wird über Verzweigungen, Schleifen und Funktio
nen gesteuert. Ob ein Block ausgeführt oder wie oft ein Block wiederholt wird, ist davon ab
hängig, ob ein Bedingungsausdruck ein <strong>true</strong> oder ein <strong>false</strong> liefert. 
</p>
<div style="border: solid 1px; padding:7px; max-width:80%">
<strong>Codeblock-Info</strong><br />
Von einem Code- oder Anweisungsblock sprechen wir, wenn mehrerer Anweisungen zu einem Block zusammengefasst werden. Anders als in vielen anderen Programmiersprachen wird ein Codeblock in Python nicht durch geschweifte Klammern {} kenntlich gemacht, sondern durch Einrückungen und einen einleitenden Doppelpunkt. 
<pre>
[if][else][for][while][try][except][finally]: 
    Anweisung1 
    Anweisung2 
    Anweisung3 
</pre>      
</div>
<h4>Fallunterscheidungen</h4>
<p>Bei Fallunterscheidungen (Verzweigungen) ist die Ausführung einer oder mehrerer Anweisungs-Blöcke davon abhängig, ob der Bedingungsausdruck den Wert <strong>true</strong> oder <strong>false</strong> liefert.</p>
<p> 
Jede Fallunterscheidung verfügt über einen <strong><u>if-Block</u></strong>: Die Anweisungen des if-Blocks, werden 
nur ausgeführt, wenn der Bedingungsausdruck den Wert <strong>true</strong> liefert.
</p>
<p>  
Der <strong><u>else-if(elif)-Block</u></strong> ist optional und wird nur ausgeführt, wenn der Bedingungsausdruck den Wert <strong>true</strong> liefert.
</p>
<p> 
Der <strong><u>else-Block</u></strong> ist optional und wird ausgeführt, wenn der Bedingungsausdruck des if-Blocks den Wert <strong>false</strong> liefert.
</p>
<pre>
alter = int(input('Wie alt bist du?'))
wahlberechtigt = 18
if(alter >= wahlberechtigt):
    print('Du bist '+str(alter)+ ' Jahre alt und damit wahlberechtigt.')
else:
    print('Du bist erst '+str(alter)+ ' Jahre alt und nicht wahlberechtigt.')   
</pre>

<h3>Schleifen</h3>
<p>Schleifen sind Kontrollstrukturen in denen ein Anweisungsblock so lange ausgeführt wird, bis 
die Laufbedingung ungültig wird, d.h. der Bedingungsausdruck den Wert false liefert.</p>   
<p>
Es gibt unterschiedliche Schleifentypen:
<ul>
    <li>Die <strong>while-Schleife</strong> wird verwendet, um einen Codeblock wiederholt auszuführen, solange eine Bedingung wahr ist.</li>
    <li>Die <strong>for-Schleife</strong> wird verwendet, um Sequenz zu durchlaufen (iterieren). Sequenzen und die for-schleife behandeln wir später im Notebook-Übungsskript Python Basics 3 <i>(sequenzen.ipynb)</i>.</li>
</ul>
Die WHILE Schleife ist einer Fallunterscheidung gar nicht so unähnlich. Dieser Schleifentyp kombiniert das Schlüsselwort WHILE mit einem Bedingungsausdruck:
<pre>
    x=1             # Die Variable x wird mit dem Wert 1 initialisiert
    max = 10        # Maximale Durchläufe 
    while x <= max: # while mit Bedingungsausdruck
        print(x)    # Ausgabe x
        x += 1      # x wird um 1 inkrementiert (mit Zuweisungsoperator +=)
</pre>

<h3>Funktionen</h3>
<p>Funktionen fassen eine Folge von Anweisungen (Befehlen) in einem Block zusammen. Das hat eine Reihe von Vorteilen:</p> 
<ul>
<li>Programme oder Programmteile können gezielt ausgeführt werden.</li> 
<li>Programme lassen sich übersichtlicher strukturieren, weil zusammengehörende Befehlsfolgen unter einem Funktionsnamen zusammengefasst werden.</li> 
<li>Programmteile, die mehrfach gebraucht werden, können immer wieder aufgerufen 
werden.</li> 
</ul>
<p><strong>Aufbau einer Funktion</strong></p>
<ul>
<li>In Python werden Funktionen durch das Schlüsselwort <strong>def</strong> eingeleitet.</li>
<li>Dem Schlüsselwort <strong>def</strong> folgt der Name der Funktion.</li>
<li>Dem Funktionsnamen folgen runde Klammern. Innerhalb dieser Klammern können ggf. Werte (Parameter) an die Funktion übergeben werden.</li>
<li>Den Klammern folgt der Funktionskörper, d.h. der Anweisungsblock.</li>
<li>Über das Schlüsselwort <strong>return</strong> kann die Funktion auch Werte zu rückliefern.</li>
</ul>
<p>
<strong>Codebeispiel: Eine personalisierte Begrüßungsfunktion </strong></ br>
<pre>
def gruss(name):                                          # Funktion
    print('Hallo '+name+'! Schön, dass Du hier bist.')

meinName = input('Gebe bitte Deinen Namen ein')
gruss(meinName)                                           # Funktionsaufruf
</pre>
<p>
Durch eine kleine Modifikation kann die Funktion auch so gestaltet werden, dass die Ausgabe des Textes nicht innerhalb der Funktion geschieht, sondern dass im Funktionskörper nur Zeichenkette zusammen
gebaut und über das Schlüsselwort <strong>return</strong> an die Stelle des Funktionsaufrufs zurückgeschickt 
wird.
<pre> 
def gruss(name):
    grusssatz = 'Hallo '+name+'! Schön, dass Du hier bist.'
    return grusssatz

meinName = input('Gebe bitte Deinen Namen ein')
gruss_zeichenkette = gruss(meinName)
print(gruss_zeichenkette)
</pre>
