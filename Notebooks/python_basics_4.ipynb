### Python Pandas

Pandas (Python Data Analysis Library) wurde speziel für die Manipulation und Analyse von Daten entwickelt. Pandas erweitert die Python Standardbibliothek um weitere Objekte (bzw. Datenstrukturen):

- **Series:** Eindimensionale, beschriftete Daten.
- **Dataframes:** Zweidimensionale, tabellenartige Daten.

Darüber hinaus stellt pandas eine Reihe nützlicher Tools zur Verfügung. Dazu zählen u.a:

- **Ein- und Ausgabe von Daten:** Das Lesen und Schreiben von CSV, JSON, Excel, und HTML Dateien. Pandas kann auch SQL und Daten direkt aus einer relationalen Datenbank lesen und schreiben.
- **Datenmanipulation:** Daten auswählen und indizieren, transformieren, bereinigen, umformen.
- **Daten aggregieren und gruppieren**
- **Fehlende Daten:** Funktionen für den Umgang mit fehlenden Daten.
- **Sowie:**: Zeitreihen, Mathematische und statistische Operationen und Datenvisualisierung.

### DataFrames
Ein DataFrame ist eine tabellartige Datenstruktur. Dataframes haben besondere Eigenschaften:

- Die Daten sind in Zeilen und Spalten organisiert (zweidimensional).
- Sowohl die Zeilen als auch die Spalten haben Beschriftungen (Labels). Die Spaltenbeschriftungen werden als Spaltennamen bezeichnet.Die Zeilenbeschriftungen bilden den Index. Dieser Index kann automatisch generiert werden (z.B. fortlaufende Zahlen) oder explizit vom Benutzer festgelegt werden.
- Die Spalten eines DataFrames können unterschiedliche Datentypen enthalten (z.B. Integer, Float, String, Boolean).
- Die Spalten und Zeilen sind veränderlich d.h. sie können hinzufügt oder entfernt werden, nachdem das DataFrame erstellt wurde.

### Von der Dictionary-Liste zum Dataframe
Um uns dem Thema DataFrame anzunähern, erinnern wir uns an das Beispiel einer Liste aus Dictionaries (Variablenname quellen). Auch diese Struktur rekonstruiert in gewisser Weise eine Tabellenstruktur. 
Die Liste der Dictionaries lässt sich aus diesem Grund leicht in einen Dataframe umwandeln.

`quellen_df = pd.DataFrame(quellen)`


```python
# Beispielcode 
quellen = [
    dict(
        idn =1338378236,
        Titel="Menschlich verständlich, aber trotzdem falsch?",
        Autor = "Kubiciel, Michael", 
        Schlagworte= ['Außenpolitik','Deutschland', 'USA']
),
    dict(
        idn = 1338378237,
        Titel="Die Welt ist nicht genug",
        Autor = "Kubiciel, Michael",
        Schlagworte= ['Außenpolitik','Deutschland', 'USA']
    ),
    dict(
        idn = 1234242478,
        Titel="Geopolitisches Delirium? US-amerikanische Denkmuster über Internet, Cyberspace und künftige Kriege",
        Autor = "Giesenfeld, Günter",
        Schlagworte= ['Geopolitik', 'Internet', 'Erster Weltkrieg', 'Außenpolitik', 'Politische Kommunikation','USA', 'Deutschland']
    )
    ]

# Die Liste Quellen in einen DataFrame umwandeln
quellen_df = pd.DataFrame(quellen)
```

### Von der CSV-Datei zum Dataframe
Excel Tabellen, aber auch CSV-Dateien stehen auch für eine Tabellenstruktur (d.h. Spalten und Zeilen). CSV-Dateien sind darauf ausgelegt, tabellarische Daten zu speichern. In dem folgenden Auszug aus einer CSV-Datei steht die erste Zeile für die Spalten(bezeichner) und jede weitere Zeile für einen Datensatz.
<pre style="background-color: #dcdcdc;border: solid 1px; padding:7px">
Nr,idn,titel,Person,Rolle,Verlag ...
0,1358939047,Trumps Gegenverfassung,"Scheppele, Kim Lane",Verfasser,Springer ...
1,1338378236,"Menschlich verständlich, aber trotzdem falsch?","Kubiciel, Michael",Verfasser,unknown ...
</pre>
Eine CSV-Datei kann in dieser Struktur schnell zu einem Dataframe gemacht werden. Pandas stellt dafür die Methode read_csv() zur Verfügung:
