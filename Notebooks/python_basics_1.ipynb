### 3. Einfache Variablen

Variablen sind Speicherbereiche für Daten. Über den Namen der Variablen (Bezeichner) kann auf den Variablenwert während des Programmlaufs zugegriffen werden.

<div style="border: solid 1px; padding:7px; max-width:80%">
<strong>Bezeichner-Info</strong><br />
Bezeichner (oder Namen) dienen dazu, Variablen, Sequenzen, Funktionen, Klassen, Module 
oder andere Objekte eindeutig zu benennen. Bezeichner sollten nur die folgenden Zeichen enthalten:

- Großbuchstaben von A bis Z.
- Kleinbuchstaben von a bis z. 
- Den Unterstrich _ (nicht an erster Stelle). 
- Zahlen von 0 bis 9. Zahlen dürfen jedoch nicht an erster Stelle stehen.
</div>

Wird einer Variable ein Startwert zugewiesen, wird diese dadurch initialisiert. Die Zuweisung eines Wertes erfolgt über das Gleichheitszeichen als Zuweisungsoperator " = ". Falls ein String (Zeichenkette) als Wert übergeben wird, muss dieser Wert in Anführungszeichen oder Hochkomma übergeben werden:  Gruss = "Hallo Welt". 
Zahlwerte werden ohne Hochkomma oder Anführungszeichen übergeben. 

`alter = 20` 

Einer existierenden Variablen kann jederzeit ein neuer Wert zugewiesen werden. Der neue Wert überschreibt in diesem Fall den alten Wert. 

`alter = 47`

Über den Print-Befehl kannst Du den Wert einer Variablen ausgeben: 

`print(alter)`

Du kannst in der Ausgabe über Print auch Text und Variablen kombinieren. Dazu verwendest Du den **f-string**. Das f vor dem String aktiviert die Verwendung von Platzhaltern in geschweiften Klammern {} innerhalb des Strings. Diese Methode ist einfacher und schneller als die ältere Methode der String-Konkatenation (Verknüpfung von String und Variablen mit dem + Zeichen)

### 4. Datentypen und Typisierung</h3>

Variablen enthalten unterschiedliche Datentypen, d.h. ein Variablenwert kann beispielsweise eine Zeichenkette, eine Fließkommazahl oder eine Ganzzahl sein. Jeder Datentyp wird in Python durch ein bestimmtes Schlüsselwort (Keyword) definiert. Wir haben überwiegend mit den folgenden Datentypen zu tun:

<table>
  <tr>
     <th>Datentyp</th>
     <th>Keyword</th>
     <th>Bescheibung</th>
     <th>Beispiel</th>
  </tr>
    <tr>
     <td>Integer</td>
     <td>int</td>
     <td>Ganzzahl</td>
     <td>1, 5, 7, 89799</td>
   </tr>
    <tr>
     <td>Float</td>
     <td>float</td>
     <td>Fließkommazahl</td>
     <td>3.99, 11.78 (Punkt als Trenner für die Nachkommastellen)</td>
   </tr>
      <tr>
     <td>String</td>
     <td>str</td>
     <td>Zeichenkette</td>
     <td>'Ich bin eine Zeichenkette.'</td>
   </tr>
</table>

Python zeichnet sich durch die dynamische Typisierung seiner Variablen aus, d.h. der Variablentyp wird erst bei der Ausführung des Programms festgelegt. Das bedeutet zweierlei: 

- Bei der Verwendung von Variablen muss der Datentyp einer Variablen nicht festgelegt werden.
- Der Datentyp einer Variablen kann sich innerhalb des Programms ändern.
- Wenn der Datentyp einer Variablen geändert werden muss, ist ein Typecasting  notwendig.
