Listen können ineinander verschachtelt werden und damit kompliziertere Formen annehmen. Listen werden durch eckige Klammern dargestellt. Die einzelnen Elemente werden durch ein Komma voneinander getrennt.</p>

##### Listen mit Schleifen durchlaufen (iterieren)

Es gibt in Python mehrere Möglichkeiten, durch eine Liste zu iterieren: 

**(1) Die einfache for-Schleife**

Die häufiste und die direkteste Methode ist die Verwendung der einfachen und schnörkelosen for-Schleife:
<pre style="background-color:#dcdcdc; border: solid 1px; padding:7px">
meine_liste = ["Apfel", "Banane", "Kirsche"]

for element in meine_liste:
    print(element)
</pre>

In diesem Beispiel nimmt die Variable element in jeder Iteration den Wert des nächsten Elements in der Liste meine_liste an.

**(2) Die Iteration über Indizes**

Wenn du auch den Index einer Liste benötigst, verwendest du die Funktion range() in Kombination mit der Länge der Liste (len().

<pre style="background-color:#dcdcdc; border: solid 1px; padding:7px">
meine_liste = ["Apfel", "Banane", "Kirsche"]

for index in range(len(meine_liste)):
    print(f"Index: {index}, Element: {meine_liste[index]}")
</pre>

**(3) Anweisungen innerhalb der for-Schleife**

Im Codeblock der Schleife kannst du die Werte deine Liste auch verändern oder andere Operationen durchführen. Im folgenden Beispiel werden die Werte einer Liste in einer for-Schleife nicht nur ausgegeben, sondern auch quadriert:

<pre style="background-color:#dcdcdc; border: solid 1px; padding:7px">
zahlen = [100, 8, 55, 26, 62, 40, 50, 62, 97, 82, 48, 37, 70, 73, 13]

for index in range(len(zahlen)):
    # Zahl quadrieren
    zahlen [index] = zahlen[index] * zahlen[index]
    # Index und Zahl ausgeben
    print(f"Index: {index}, Element: {zahlen[index]}")
</pre>

**(4) Die while-Schleife**

Die while-Schleife funktioniert auch, ist aber nicht die erste Wahl, weil für den Bedingunsausdruck zusätzlich ein Zähler (hier die Variable index) definiert und im Schleifenkörper inkrementiert (hochgezählt) werden muss.

<pre style="background-color:#dcdcdc; border: solid 1px; padding:7px">
index = 0
while index < len(zahlen):
    print(f"Index: {index} Wert: {zahlen[index]}")
    zahlen [index] = zahlen[index] * 2
    index += 1
</pre>

##### Tupeln
Im Unterschied zu Listen hanelt es sich bei Tupeln Sequenzen, deren Elemente nicht manipuliert werden können. Im Unterschied zur Liste ist es hier nicht möglich, Elemente hinzuzufügen, zu entfernen oder zu bearbeiten.

Tupel werden mit runden Klammern deklariert. 

`tupel = (10,23,27,29)`

Lesefunktionen, wie len(), max(), min() etc., können auch auf Tupel angewandt werden, weil diese das Tupel nicht verändern.

##### Dictionaries

In einer Liste oder einem Tupel können zusammenhängende Daten sehr schön unter einem Bezeichner zusammengefasst werden. Das Problem besteht zumeist darin, dass
man immer genau wissen muss, unter welchem Index ein bestimmter Inhalt gespeichert ist. 

Ein Dictionary hat den Vorteil, dass Werte unter einem Schlüsselwort gespeichert werden. Diese Schlüsselwort kann Auskunft über die Beschaffenheit der gespeicherten Werte.

Ein Dictionary besteht aus Schlüssel – Wert Paaren (Items), die durch ein Komma abgetrennt, innerhalb von geschweiften Klammern stehen. 

<pre style="background-color:#dcdcdc; border: solid 1px; padding:7px">
    person = {
    'Vorname': 'Uschi',
    'Nachname': 'Krawutttke',
    'Alter': 47
    }
    </pre>

Dictionary items sind geordnet und veränderbar. Ein Schlüssel darf in einem Dictionary aber nur einmal vorkommen.

- Um ein Dictionary anzulegen, kann ich die key-value Paare wie oben notieren oder den Konstruktor verwenden: `person = dict(Vorname = "Uschi", Nachname = 'Krawuttke', Alter = 47)`

- mit **update()** kann ich mehrere Werte gleichzeitig aktualisieren und hinzufügen:`person.update({'Hobbies': ['Radeln', 'Joggen', 'Golfen'],'Beruf': 'Ingenieurin'})`

- Der Zugriff auf den Wert eines bestimmtes Item erfolgt über den Bezeichner und den Schlüssel: `print(person['Nachname'])`. Der Wert eines Items lässt sich auf diese Weise nicht nur abrufen, sondern auch ändern: `person['Nachname'] = 'Krawutttke-Schnitzel'`

- Um ein Item zu löschen, verwenden wir die Methode **pop()**: `person.pop('Hobbies')`

- Ich kann dem Dictionary Items hinzufügen: `person["Hobbies"] = ['Schwimmen', 'Radfahren', 'Lesen']`. Die Value-Werte können beliebigen Datentyps sein. Das schließt auch Objekttypen wie Listen, Tupeln oder andere Dictionaries (nested Dictionaries) ein. In diesem Fall wird dem Item 'Hobbies' eine Liste hinzugefügt.

- mit der **for-Schleife** kannst du ein Dictionary durchlaufen (iterieren):
<pre style="background-color:#dcdcdc; border: solid 1px; padding:7px">
    for x in person:
        print(f"{x}: {person[x]}")
</pre>

````python
# Codeblock für die Aufgabe 5
quellen = [
    dict(
    idn =1338378236,
    Titel="Menschlich verständlich, aber trotzdem falsch?",
    Autor = "Kubiciel, Michael", 
    Schlagworte= ['Außenpolitik','Deutschland', 'USA']
),
    dict(
        idn = 1338378237,
        Titel="Die Welt ist nicht genug",
        Autor = "Kubiciel, Michael",
        Schlagworte= ['Außenpolitik','Deutschland', 'USA']
    ),
    dict(
        idn = 1234242478,
        Titel="Geopolitisches Delirium? US-amerikanische Denkmuster über Internet, Cyberspace und künftige Kriege",
        Autor = "Giesenfeld, Günter",
        Schlagworte= ['Geopolitik', 'Internet', 'Erster Weltkrieg', 'Außenpolitik', 'Politische Kommunikation','USA', 'Deutschland']
    )
    ]

for i in range(len(quellen)):
    for x in quellen[i]:
        if x == "Schlagworte":
            for j in range(len(quellen[i][x])):
                print(f"Schlagwort {j+1}: {quellen[i][x][j]}")
        else:
            print(x, quellen[i][x])

    print("-------------------")
````
